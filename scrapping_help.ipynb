```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.siemens-energy.com/global/en/home/products-services/product-offerings.html"  # Replace with the URL of the webpage you want to scrape
response = requests.get(url)

if response.status_code == 200:
    print("Page successfully loaded")
else:
    print(f"Failed to retrieve the webpage, status code: {response.status_code}")
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
headings = soup.find_all()
for heading in headings:
    print(heading.text.strip())  # Clean up extra spaces
```

```python
product_elements = soup.find('div', class_='custom-container')  # Adjust the class name based on inspection
```

```python
text = product_elements.get_text(separator='\n', strip=True) if product_elements else "No text found"
```
