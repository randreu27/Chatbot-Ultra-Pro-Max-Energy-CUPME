```python
import fitz  # PyMuPDF

def extraer_texto_pdf(ruta_pdf):
    doc = fitz.open(ruta_pdf)
    texto = ""
    for pagina in doc:
        texto += pagina.get_text("text") + "\n"
    return texto

# Cargar documento
ruta_pdf = "Declaración Universal de los Derechos Humanos.pdf"
documento = extraer_texto_pdf(ruta_pdf)

print(documento[:500])  # Ver parte del texto extraído
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

# Modelo pequeño para embeddings
modelo_embeddings = SentenceTransformer("all-MiniLM-L6-v2")

# Dividir el texto en fragmentos
fragmentos = documento.split(". ")  # Separar por oraciones

# Convertir cada fragmento en embeddings
embeddings = modelo_embeddings.encode(fragmentos)

# Crear base de datos FAISS
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
fragmentos
```

```python
def retrieve(query, model, index, fragments):
    query_embedding = model.encode([query])
    D, I = index.search(np.array(query_embedding), k=1)  # Obtener el más cercano
    print(I)
    return fragments[I[0][0]]  # Devolver el fragmento más relevante

query = "Toda persona tiene derecho a participar en el gobierno de su país?"
fragmento_relevante = retrieve(query, modelo_embeddings, index, fragmentos)

print("Documento relevante:", fragmento_relevante)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Modelo pequeño para generación
modelo_generacion = "tiiuae/falcon-rw-1b"  # También puedes probar "facebook/opt-350m"
tokenizer = AutoTokenizer.from_pretrained(modelo_generacion)
modelo = AutoModelForCausalLM.from_pretrained(modelo_generacion)

# Generar respuesta basada en el fragmento
input_text = fragmento_relevante + "\nPregunta: " + query + "\nRespuesta:"
inputs = tokenizer(input_text, return_tensors="pt")

output = modelo.generate(**inputs, max_length=100)
respuesta = tokenizer.decode(output[0], skip_special_tokens=True)

print("Respuesta del chatbot:", respuesta)
```

```python
# Generar respuesta basada en el fragmento
input_text = fragmento_relevante + "\nPregunta: " + query + "\nRespuesta:"
inputs = tokenizer(input_text, return_tensors="pt")

output = modelo.generate(
    **inputs,  
    max_length=150,  
    do_sample=True,  # Activa muestreo en vez de tomar siempre la mejor opción  
    temperature=0.7,  # Controla la aleatoriedad (valores más altos = más diversidad)  
    top_p=0.9,  # Núcleo de muestreo (descarta opciones de baja probabilidad)  
    top_k=50  # Considera solo las 50 opciones más probables  
)
respuesta = tokenizer.decode(output[0], skip_special_tokens=True)

print("Respuesta del chatbot:", respuesta)
```
